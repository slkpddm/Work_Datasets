```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(20) #if we give seed means data will be same for all
df=pd.DataFrame(np.random.randn(10,4),index=pd.date_range('2022-03-14',periods=10),columns=['a','b','c','d'])
```

```python
df
```

```python
# If we add column e which is a result of addition of columns c and d
df["e"]=df['c']+df['d']
df
```

```python
#If we find square root of column d
df["f"]=np.sqrt(df['d'])
df
```

```python
#We can also find the square of d
df['g']=df['d'].apply(lambda x : x**2)
df
```

```python
# We can also find log
df['h']=df['d'].apply(lambda x :np.log(x))
df
```

```python
# We can also apply the function
def test(x):
    if (x<0):
        return x
    else:
        return np.log(x)
```

```python
df['j']=df['d'].apply(test)
df
```

```python
df.index
```

```python
#Now we add the new columns with date 
df['new_date']=df.index
df
```

```python
# We can find the data types
df.dtypes
```

```python
# If I want seperate date , month and year seperately into seperate columns
df["date"]=df.index.day
df["month"]=df.index.month
df["year"]=df.index.year
```

```python
df
```

```python
df.dtypes
```

```python
df["week"]=df.index.day_of_week
df["week_name"]=df.index.day_name()
df
```

```python
df.groupby(by="week_name").count() # We can make group every column for count
```
